```python
import yfinance as yf
import pandas as pd
```

```python
#Using the yfinance library to extract stock data for tesla
tesla = yf.Ticker("TSLA")
```

```python
# tesla stock info
tesla_info=tesla.info
tesla_info
```

```python
#tesla country info
tesla_info['country']
```

```python
#tesla data
tesla_data = tesla.history(period="max")
tesla_data.head()
```

```python
#tesla revenue
tesla_revenue = tesla.history(period="max")
tesla_revenue.tail()
```

```python
#Make graph for Tesla
tesla_data.reset_index(inplace=True)
tesla_data.plot(x="Date", y="Open")
```
