```python
#Basic Plotting
import matplotlib.pyplot as plt
country = ['China', 'India', 'United Sates', 'Indonesia', 'Pakistan', 'Brazil']
population = [1439323776, 1380004385, 331002651, 273523615, 220892340, 212559417]
plt.bar(country,population)
plt.title('Population of Countries (2020)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
#making an horizontal bar graph
plt.barh(country,population)
plt.title('Population of Countries (2020)')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```

```python
#making a line plot
year = [1851, 1861, 1871, 1881, 1891,
        1901, 1911, 1921, 1931, 1941,
        1951, 1961, 1971, 1981, 1991,
        2001, 2011]
population = [ 27368800, 28917900, 31484700, 34934500, 37802400,
               41538200, 42189800, 43904100, 46073600, 44870400,
               50286900, 52807400, 55928000, 56357500, 57438700,
               59113016, 63285145 ]
plt.plot(year, population)
plt.title('Population of the United Kingdom')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
#making boxplots
import pandas as pd
url = 'https://raw.githubusercontent.com/fivethirtyeight/data/master/births/US_births_2000-2014_SSA.csv'
mydata = pd.read_csv(url)
plt.boxplot(mydata.births, vert=False)
plt.title('Distribution of Number of Daily Births in the USA (2000-2014)')
plt.xlabel('Number of Births')
plt.show()
```

```python
#making vertical boxplots
plt.boxplot(mydata.births, vert=True)
plt.title('Distribution of Number of Daily Births in the USA (2000-2014)')
plt.ylabel('Number of Births')
plt.show()
```

```python
#making a scatter plot
temperature = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
icecreamsales = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]
plt.scatter(temperature, icecreamsales)
plt.title('Ice cream sales vs Temperature')
plt.xlabel('Temperature (degrees C)')
plt.ylabel('Ice cream sales')
plt.show()
```

```python
#make bar chart of top 10 world's tallest buildings
Building = ['Burj Khailfa', 'Merdeka 118', 'Shanghai Tower', 'Abraj Al-bait Clock Tower', 'Ping An International Finance center',
            'Lotte World Center', 'One World Trade Center', 'Guangzhou CTF Finance Center', 'Tianjin CTF Finance Center', 'China Zun']
Height = [828, 678.9, 632, 601, 599.1, 554.5, 541.3, 530, 530, 527.7]
plt.bar(Building, Height)
plt.title('Top 10 Tallest Buildings in the World (2022)')
plt.xlabel('Buildings')
plt.ylabel('Height (m)')
plt.show()
```

```python
#make bar chart more tidy by increase chart size
plt.figure(figsize=(15,10))
plt.bar(Building, Height)
plt.title('Top 10 Tallest Buildings in the World (2022)')
plt.xlabel('Buildings')
plt.ylabel('Height (m)')
plt.xticks(rotation=90)
plt.show()
```
